```python
import pandas as pd
import csv
import numpy as np
import requests
import matplotlib.pyplot as plt
import json
from scipy.stats import linregress
```

```python
kindle_csv = "C:/Users/freak/Downloads/kindle_reviews.csv"
kindle_data = pd.read_csv(kindle_csv)
kindle_data
```

```python
kindle_drop = kindle_data.dropna()
```

```python
word_list = []

for review in kindle_drop["reviewText"]:
    review_replace = review.replace(".", " ")
    review_replace = review.replace("&#34;", " ")
    review_replace = review_replace.replace("*", " ")
    review_replace = review_replace.replace("(", " ")
    review_replace = review_replace.replace(")", " ")
    review_replace = review_replace.replace(",", " ")
    review_replace = review_replace.replace('''"''', " ")
    review_replace = review_replace.replace("?", " ")
    review_replace = review_replace.replace("!", " ")
    review_replace = review_replace.replace("/", " ")
    review_replace = review_replace.replace(":", " ")
    
    word_split = review.split(" ")
    for word in word_split:
        word_lower = word.lower()
        word_list.append(word_lower)

len(word_list)
```

```python
word_df = pd.DataFrame(word_list)

word_df
```
